```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Vehicle Type': ['Sedan', 'SUV', 'Truck', 'Sedan', 'SUV'],
    'Sales': [10000, 12000, 8000, 7000, 9000]  # Example sales figures
}

# Creating DataFrame
sales_data = pd.DataFrame(data)

# Filter data for recession periods (e.g., years 2020 and 2021)
recession_years = [2020, 2021]
recession_sales_data = sales_data[sales_data['Year'].isin(recession_years)]

# Plotting separate lines for each vehicle type
for vehicle_type, group_data in recession_sales_data.groupby('Vehicle Type'):
    plt.plot(group_data['Year'], group_data['Sales'], marker='o', linestyle='-', label=vehicle_type)

# Adding title and labels
plt.title('Sales Trends of Different Vehicle Types During Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Vehicle Type': ['Sedan', 'SUV', 'Truck', 'Sedan', 'SUV'],
    'Sales': [10000, 12000, 8000, 7000, 9000]  # Example sales figures
}

# Creating DataFrame
sales_data = pd.DataFrame(data)

# Define recession period and non-recession period
recession_years = [2020, 2021]
non_recession_years = [2018, 2019, 2022]

# Filter data for recession and non-recession periods
recession_sales_data = sales_data[sales_data['Year'].isin(recession_years)]
non_recession_sales_data = sales_data[sales_data['Year'].isin(non_recession_years)]

# Combine recession and non-recession data
combined_sales_data = pd.concat([recession_sales_data.assign(Period='Recession'), 
                                 non_recession_sales_data.assign(Period='Non-Recession')])

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_sales_data, x='Year', y='Sales', hue='Vehicle Type', style='Period', marker='o')
plt.title('Comparison of Sales Trend per Vehicle Type for Recession and Non-Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample GDP data
years = [2018, 2019, 2020, 2021, 2022]
gdp_recession = [2000, 2100, 1800, 1700, 1900]  # Example GDP during recession period
gdp_non_recession = [2200, 2300, 2400, 2500, 2600]  # Example GDP during non-recession period

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot GDP during recession period
axs[0].plot(years, gdp_recession, marker='o', linestyle='-', color='r')
axs[0].set_title('GDP Variation During Recession Period')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('GDP')
axs[0].grid(True)

# Plot GDP during non-recession period
axs[1].plot(years, gdp_non_recession, marker='o', linestyle='-', color='b')
axs[1].set_title('GDP Variation During Non-Recession Period')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('GDP')
axs[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
months = np.arange(1, 13)
sales = np.array([500, 600, 700, 900, 1100, 1300, 1500, 1400, 1200, 1000, 800, 600])  # Example sales figures
seasonality = np.array([0.5, 0.6, 0.7, 0.9, 1.0, 1.2, 1.5, 1.4, 1.2, 1.0, 0.8, 0.6])  # Example seasonality factors

# Calculate bubble sizes proportional to sales
bubble_sizes = sales / 100  # Adjust the scaling factor as needed

# Plot the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(months, sales, s=bubble_sizes*1000, c=seasonality, cmap='coolwarm', alpha=0.7, edgecolors='k', linewidths=1)
plt.colorbar(label='Seasonality')
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(months)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2018, 2019, 2020, 2021, 2022]
sales_volume = [10000, 12000, 8000, 7000, 9000]  # Example sales volume during recession period
avg_vehicle_price = [25000, 26000, 24000, 23000, 25000]  # Example average vehicle price during recession period
recession_indicator = [1, 0, 1, 1, 0]  # Example recession indicator (1 for recession, 0 for non-recession)

# Filter data for recession periods
recession_sales_volume = [sales_volume[i] for i in range(len(years)) if recession_indicator[i] == 1]
recession_avg_vehicle_price = [avg_vehicle_price[i] for i in range(len(years)) if recession_indicator[i] == 1]

# Plot scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(recession_avg_vehicle_price, recession_sales_volume, color='b', alpha=0.7)
plt.title('Correlation Between Average Vehicle Price and Sales Volume During Recessions')
plt.xlabel('Average Vehicle Price ($)')
plt.ylabel('Sales Volume')
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2018, 2019, 2020, 2021, 2022]
vehicle_types = ['Sedan', 'SUV', 'Truck']
sales = {
    'Sedan': [10000, 12000, 8000, 7000, 9000],
    'SUV': [8000, 9000, 6000, 5000, 7000],
    'Truck': [5000, 6000, 4000, 3000, 5000]
}
unemployment_rate = [5.2, 5.5, 7.8, 9.2, 8.5]  # Example unemployment rate during recession period

# Plot line plot
plt.figure(figsize=(10, 6))
for vehicle_type in vehicle_types:
    plt.plot(years, sales[vehicle_type], marker='o', label=vehicle_type)

# Plot unemployment rate as a secondary y-axis
plt.twinx()
plt.plot(years, unemployment_rate, linestyle='--', color='red', marker='s', label='Unemployment Rate')

# Add title and labels
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales During Recession Period')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)

# Show legend
plt.legend(loc='upper left')

# Show plot
plt.show()
```

```python
import dash
from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go

# Sample data for recession report statistics
years = [2018, 2019, 2020, 2021, 2022]
sales = [10000, 12000, 8000, 7000, 9000]
unemployment_rate = [5.2, 5.5, 7.8, 9.2, 8.5]

# Initialize the Dash app
app = dash.Dash(__name__)

# Define dropdown options
vehicle_types = [
    {'label': 'Sedan', 'value': 'sedan'},
    {'label': 'SUV', 'value': 'suv'},
    {'label': 'Truck', 'value': 'truck'}
]

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Automobile Sales Dashboard'),  # Title of the dashboard
    
    # Dropdown for selecting vehicle type
    html.Label('Select Vehicle Type'),
    dcc.Dropdown(
        id='vehicle-dropdown',
        options=vehicle_types,
        value='sedan'
    ),
    
    # Division for output display
    html.Div(id='output-container', className='output-container'),
    
    # Graphs for recession report statistics
    html.Div(children=[
        dcc.Graph(
            id='sales-graph',
            figure={
                'data': [
                    {'x': years, 'y': sales, 'type': 'line', 'name': 'Sales'}
                ],
                'layout': {
                    'title': 'Automobile Sales During Recession Period',
                    'xaxis': {'title': 'Year'},
                    'yaxis': {'title': 'Sales'}
                }
            }
        ),
        dcc.Graph(
            id='unemployment-graph',
            figure={
                'data': [
                    {'x': years, 'y': unemployment_rate, 'type': 'line', 'name': 'Unemployment Rate', 'line': {'color': 'red'}}
                ],
                'layout': {
                    'title': 'Unemployment Rate During Recession Period',
                    'xaxis': {'title': 'Year'},
                    'yaxis': {'title': 'Unemployment Rate'}
                }
            }
        )
    ])
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
